## How many votes has each type of resturant received from the customers?


```python
zomatoproject.head()
```

```python
grouped_data = zomatoproject.groupby('listed_in(type)')['votes'].sum() # created with group to the drow in the graphs
result = pd.DataFrame({'votes': grouped_data})   # create the new dataframe to the votes and add in the group data
plt.plot(result,c="green",marker="o")  # plot the data in the line graph
plt.xlabel("Type of resturant",c="red",size=20) # name of label,coloer,and size
plt.ylabel("votes",c="red",size=20)
```

# CONCLUSION : Dining resturants has recieved maximum vote and buffet in mimimum vote

## Which mode (online or offline) has received the maximum rating ?

```python
plt.figure(figsize = (6,6))
sns.boxplot(x="online_order",y= 'rate', data=zomatoproject)
# figure is use to the size of the graph depents
# boxplot is use to the graph drow in the box type
```

 # CONCLUSION: offline received lower rating in comparison to online order

# which type of order resturants received more offline order so that Zomato can process to customers with some good offers?

```python
# heatplot
# created the pivot tables to the main tables
pivot_table = zomatoproject.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True, cmap='YlGnBu',fmt='d') # heatmap:graph type, aanot:,colur map: YL.GRN.bule, fmt:formet
plt.title("Heatmap")
plt.xlabel("online order")
plt.ylabel("Listed In (Type)")
plt.show()
print(pivot_table)
```

# CONCLUSION: Dining restaurants primarilly accept  offline orders whereas cafes primarily receive online orders. This suggests that clients prefer orders in person at resturants but prefer online ordering at cafes.
